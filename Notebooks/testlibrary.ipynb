```python
from pDistinguisher import knn_distinguish
import pandas as pd
```

```python
print("--------Iris---------")
# data_lst = []
classes = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
df_origin = pd.read_csv("../data/Iris/Iris-Original", header=None, sep=',')
df_oo = pd.read_csv("../data/Iris/Iris-test", header=None, sep=',')
last_col = df_origin.iloc[:, -1]

# Select all but the last column and move it to the first position
df_new = pd.concat([last_col, df_origin.iloc[:, :-1]], axis=1)
# print(df_new)
# print(df_new)
# print(df_new)
# test_SP = knn_distinguish(df_new, df_oo, classes)
train_iris = knn_distinguish(df_new.iloc[:-1, ], classes)
sig_function, popt, _ =train_iris.fit()
```

```python
train_iris.predict(df_oo, sig_function, popt)
```

```python
print("-----------------test heart-----------------")
df_train = pd.read_csv("../data/SPECT/SPECTF.train", header=None, sep=',')
class_heart = ['1', '0']
df_train[0] = df_train[0].astype(str)
train_heart = knn_distinguish(df_train, class_heart)
sig, popt, _ = train_heart.fit()
```

```python
df_test = pd.read_csv("../data/SPECT/test.csv", header=None, sep=', ')
print(df_test)
train_heart.predict(df_test, sig, popt)
```
